```python
tickers = [
    "AAPL",  # Apple Inc.
    "MSFT",  # Microsoft Corporation
    "GOOGL", # Alphabet Inc.
    "AMZN",  # Amazon.com Inc.
    "META",  # Meta Platforms Inc.
    "NVDA",  # NVIDIA Corporation
    "TSLA",  # Tesla Inc.
    "ADBE",  # Adobe Inc.
    "CRM",   # Salesforce.com Inc.
    "NFLX",  # Netflix Inc.
    "INTC",  # Intel Corporation
    "CSCO",  # Cisco Systems Inc.
    "IBM",   # International Business Machines Corporation
    "ORCL",  # Oracle Corporation
    "QCOM",  # Qualcomm Incorporated
    "PYPL",  # PayPal Holdings Inc.
    "SNAP",  # Snap Inc.
    "TWTR",  # Twitter Inc.
    "ZM",    # Zoom Video Communications Inc.
    "ATVI"   # Activision Blizzard Inc.
]
```

```python
startdate = "2021-01-01"
enddate = "2024-01-01"

cointegrated_pairs = []

data = pd.DataFrame()

for ticker in tickers:
    try:
        # Download the Adjusted Close prices for each ticker
        ticker_data = yf.download(ticker, start=startdate, end=enddate)['Adj Close']
        
        if not ticker_data.empty:
            # Add the series to the DataFrame with the ticker as the column name
            data[ticker] = ticker_data
        else:
            print(f"No data for {ticker}")
    except Exception as e:
        print(f"Error downloading {ticker}: {e}")


```

```python
fintickers = data.columns
numtickers = len(data.columns)
alpha = 0.05


# Two-step cointegration test
def two_step_coint_test(stock1, stock2):
    # Fit OLS regression of stock1 on stock2
    model = OLS(stock1, stock2).fit()
    residuals = model.resid
    
    # Perform the Augmented Dickey-Fuller test on residuals
    adf = adfuller(residuals)
    
    # Return p-value of the ADF test
    return adf[1]

# Iterate through combinations of stocks to test for cointegration
for i in range(numtickers):
    for j in range(i+1, numtickers):
        stock1 = data[fintickers[i]].copy()
        stock2 = data[fintickers[j]].copy()
        
        # Remove NaN and Inf values from both stock1 and stock2
        combined = pd.concat([stock1, stock2], axis=1)
        combined = combined.replace([np.inf, -np.inf], np.nan).dropna()

        if not combined.empty:  # Ensure we have data after cleaning
            stock1_clean = combined.iloc[:, 0]
            stock2_clean = combined.iloc[:, 1]
            
            # Apply the two-step cointegration test
            p_val = two_step_coint_test(stock1_clean, stock2_clean)

            if p_val < alpha:
                cointegrated_pairs.append((fintickers[i], fintickers[j]))

cointegrated_pairs
```

```python
spreads = pd.DataFrame()


for pair in cointegrated_pairs:
    tk1, tk2 = pair
    spreads[f"{tk1} - {tk2}"] = data[tk1] - data[tk2]
    

spreads.head()
```

```python
spreadsdict = {
    'Mean Difference' : spreads.mean()
}

mean_spread = pd.DataFrame(spreadsdict)

mean_spread
```
